```python
import json
import os
import shutil

# Directories
json_dir = '/Users/mehmetkarabulut/Library/Mobile Documents/com~apple~CloudDocs/work/drone/uavdt-DatasetNinja/test/ann'
image_dir = '/Users/mehmetkarabulut/Library/Mobile Documents/com~apple~CloudDocs/work/drone/uavdt-DatasetNinja/test/img'
txt_dir = '/Users/mehmetkarabulut/Library/Mobile Documents/com~apple~CloudDocs/work/drone/uavdt-DatasetNinja/test/test_txt'
birdsview_img_dir = '/Users/mehmetkarabulut/Library/Mobile Documents/com~apple~CloudDocs/work/drone/uavdt-DatasetNinja/test/test_birdsview_img'

# Ensure the target directories exist
os.makedirs(txt_dir, exist_ok=True)
os.makedirs(birdsview_img_dir, exist_ok=True)

# Function to check if "birds view" label exists
def is_birds_view(tags):
    for tag in tags:
        if tag['name'] == 'bird view':
            return True
    return False

# Function to convert JSON annotations to YOLO format
def convert_to_yolo(json_file, txt_dir):
    with open(json_file, 'r') as f:
        data = json.load(f)

    # Extract image dimensions
    image_width = data['size']['width']
    image_height = data['size']['height']

    # Prepare the YOLO format data
    yolo_data = []
    for obj in data['objects']:
        class_id = obj['classId']  # You might need a mapping from classId to YOLO class
        xmin, ymin = obj['points']['exterior'][0]
        xmax, ymax = obj['points']['exterior'][1]

        # Calculate the center coordinates, width, and height
        x_center = (xmin + xmax) / 2 / image_width
        y_center = (ymin + ymax) / 2 / image_height
        width = (xmax - xmin) / image_width
        height = (ymax - ymin) / image_height

        yolo_data.append(f"{class_id} {x_center} {y_center} {width} {height}")

    # Determine the name of the output file
    base_filename = os.path.splitext(os.path.basename(json_file))[0]
    txt_file = os.path.join(txt_dir, f"{base_filename}.txt")

    # Write to the YOLO format text file
    with open(txt_file, 'w') as f:
        for line in yolo_data:
            f.write(line + '\n')

    print(f"YOLO annotations saved to {txt_file}")

# Iterate through JSON files in the specified directory
for filename in os.listdir(json_dir):
    if filename.endswith('.json'):
        json_file = os.path.join(json_dir, filename)
        with open(json_file, 'r') as f:
            data = json.load(f)
            # Check if the JSON file contains the "birds view" label
            if is_birds_view(data['tags']):
                # Convert annotations to YOLO format
                convert_to_yolo(json_file, txt_dir)
                
                # Copy the image to the birdsview_img directory
                image_filename = os.path.splitext(filename)[0]
                source_image_path = os.path.join(image_dir, image_filename)
                destination_image_path = os.path.join(birdsview_img_dir, image_filename)
                
                shutil.copy(source_image_path, destination_image_path)
                print(f"Image {image_filename} copied to {birdsview_img_dir}")
```
